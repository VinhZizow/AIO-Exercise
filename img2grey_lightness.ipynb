```python
import matplotlib.image as mping
import numpy as np
import matplotlib.pyplot as plt
img = mping.imread('/content/dog.jpeg')
img
```

```python
#((max(rgb) + min(rgb)) /2
def color2gray(vector):
  return np.max(vector)*0.5 + np.min(vector)*0.5

gray_01 = np.apply_along_axis(color2gray, axis= 2, arr= img)
gray_01.shape
```

```python
plt.imshow(gray_01, cmap=plt.get_cmap('gray'))
plt.show()
gray_01[0,0]
```
